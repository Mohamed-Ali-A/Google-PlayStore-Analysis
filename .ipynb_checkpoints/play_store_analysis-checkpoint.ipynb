## Overview

Bottom Line Up Front: The Google Play Store hosts millions of applications
across dozens of categories. This analysis identifies which app categories,
pricing models, and content ratings drive the highest installs and ratings —
giving our startup actionable intelligence before entering the market.

Summary: Using a dataset of 2.3M+ Play Store apps, we clean and explore
key attributes including category, rating, installs, size, price, and content
rating. Initial findings suggest that free apps dominate install counts,
a small number of categories capture the majority of traffic, and user
ratings cluster tightly between 4.0–4.5 stars. These patterns directly
inform where a new entrant can realistically compete and gain early traction.

Story: A startup wants to launch its first Android app. Before writing
a single line of code, they need to know — where do users actually spend
their time, and what does a successful app look like? This analysis
answers that question with data.

### Columns to Drop or Deprioritize
- `App Name`, `Developer ID`, `Developer Website`, `Developer Email`
  — identifiers, not analytical dimensions
- `Scraped Time` — metadata, not business-relevant

### Known Data Quality Issues to Address
- `Installs` column contains strings like `"1,000,000+"` — needs numeric conversion
- `Size` contains mixed formats (`"19M"`, `"Varies with device"`) — needs standardization
- `Rating` may contain nulls for apps with no reviews — needs handling
- `Price` may contain `"$"` prefix — needs stripping and casting to float
- `Released` and `Last Updated` are strings — need date parsing
- Potential duplicate app entries (same app listed multiple times)
- Null values likely across Rating, Rating Count, Size, and Installs

## Data Preparation



```python
# Data Prep Code Here
# ============================================================
# STEP 4: DATA PREPARATION & CLEANING
# ============================================================

from pyspark.sql.functions import regexp_replace, trim, when

# Clean Installs — remove + and ,
df_clean = df.withColumn(
    "Installs_Clean",
    regexp_replace(col("Installs"), "[+,]", "").cast("long")
)

# Clean Price — remove $ sign
df_clean = df_clean.withColumn(
    "Price_Clean",
    regexp_replace(col("Price"), "\\$", "").cast("float")
)

# Verify
df_clean.select("Installs", "Installs_Clean", "Price", "Price_Clean").show(10)
```

```python
# Standardize Type column — fix "0" and "NaN" values
df_clean = df_clean.withColumn(
    "Type",
    when(col("Type") == "0", "Free")
    .when(col("Type") == "NaN", None)
    .otherwise(col("Type"))
)

df_clean.groupBy("Type").count().show()
```

```python
# Drop rows where Rating is null
# Drop rows where Type is null
# Drop rows where Content Rating is null

df_clean = df_clean.dropna(subset=["Rating", "Type", "Content Rating"])

print(f"Rows after dropping nulls: {df_clean.count():,}")
```

```python
# Cast Reviews to integer
df_clean = df_clean.withColumn(
    "Reviews_Clean",
    col("Reviews").cast("long")
)

# Flag Varies with device in Size
df_clean = df_clean.withColumn(
    "Size_Clean",
    when(col("Size") == "Varies with device", None)
    .when(col("Size").endswith("M"),
        regexp_replace(col("Size"), "M", "").cast("float"))
    .when(col("Size").endswith("k"),
        (regexp_replace(col("Size"), "k", "").cast("float") / 1024))
    .otherwise(None)
)

df_clean.select("Size", "Size_Clean").show(10)
```

```python
# Drop full duplicate rows
df_clean = df_clean.dropDuplicates()

# Keep only one row per App — the one with most reviews
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

window = Window.partitionBy("App").orderBy(col("Reviews_Clean").desc())
df_clean = df_clean.withColumn("row_num", row_number().over(window)) \
                   .filter(col("row_num") == 1) \
                   .drop("row_num")

print(f"Final row count after deduplication: {df_clean.count():,}")
```

```python
# Select only relevant columns for Tableau
df_final = df_clean.select(
    "App",
    "Category",
    "Rating",
    "Reviews_Clean",
    "Size_Clean",
    "Installs_Clean",
    "Type",
    "Price_Clean",
    "Content Rating",
    "Genres",
    "Last Updated"
)

# Rename columns cleanly
df_final = df_final.toDF(
    "App", "Category", "Rating", "Reviews",
    "Size_MB", "Installs", "Type", "Price",
    "Content_Rating", "Genres", "Last_Updated"
)

# Export to CSV via Pandas
df_final.toPandas().to_csv("google_playstore_cleaned.csv", index=False)

print("Cleaned file saved as google_playstore_cleaned.csv ✓")
print(f"Final shape: {df_final.count():,} rows x {len(df_final.columns)} columns")
df_final.show(5)
```
